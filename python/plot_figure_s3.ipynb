```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from utils import sigmoid
```

```python
f=np.load('../output/sims/snr_threshold/sim_results.npz')
snrs=f['snrs']
snr_thresholds=f['snr_thresholds']
```

```python
plt.rcParams.update({
    'font.size': 14,
    'axes.spines.right': False,
    'axes.spines.top': False,
    'xtick.major.size': 6,
    'xtick.major.width': 1.2,
    'ytick.major.size': 6,
    'ytick.major.width': 1.2,
    'legend.frameon': False,
    'legend.handletextpad': 0.1,
    'svg.fonttype': 'none',
    'text.usetex': False
})

mean_vals = np.mean(np.log(snr_thresholds), axis=-1)
std_vals = np.std(np.log(snr_thresholds), axis=-1)

fig=plt.figure()
plt.plot(snrs, mean_vals, label='log(threshold)')
plt.fill_between(snrs, mean_vals - std_vals, mean_vals + std_vals, alpha=0.3)
plt.xlabel('SNR (dB)')
plt.ylabel('log(threshold)')
fig.savefig('../output/figures/figure_s3.pdf')
```
